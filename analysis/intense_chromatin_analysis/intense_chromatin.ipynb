```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from statannotations.Annotator import Annotator

fontsize = 13

plt.rcParams.update(
    {
        "font.family": "sans-serif",
        "font.sans-serif": ["Arial"],
        "axes.labelsize": fontsize,
        "legend.fontsize": fontsize,
        "xtick.labelsize": fontsize,
        "ytick.labelsize": fontsize,
        "axes.titlesize": fontsize,
        "legend.handlelength": 2,
    }
)

pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df_total = pd.read_csv("../../data/intense_chromatin/intense_chromatin_total.csv")
df_total = df_total.drop("Unnamed: 0", axis=1)


df = pd.read_csv("../../data/intense_chromatin/intense_chromatin_all.csv")
df = df.drop("Unnamed: 0", axis=1)
```

```python
# Specify width and height in millimeters
fig_width_mm = 210  # width of A4 page
fig_height_mm = 75

# Convert millimeters to inches (1 inch = 25.4 mm)
fig_width_inches = fig_width_mm / 25.4
fig_height_inches = fig_height_mm / 25.4

fig, axes = plt.subplots(
    nrows=1, ncols=2, figsize=(fig_width_inches, fig_height_inches), tight_layout=True, dpi=300, sharey=False
)

# Define labels and color scheme
labels = ["Non-Confined", "Confined"]
my_pal = {labels[0]: "royalblue", labels[1]: "red"}


feature = "Major axis intense chromatin (um)"
sns.boxplot(
    data=df,
    x="Condition",
    y=feature,
    order=labels,
    palette=my_pal,
    # color='white',
    width=0.5,
    boxprops=dict(edgecolor="black"),
    capprops=dict(color="black"),
    whiskerprops=dict(color="black"),
    medianprops=dict(color="black"),
    # flierprops=dict(marker='o', markersize=4, markerfacecolor='white', markeredgecolor='white'),
    ax=axes[0],
)

axes[0].set(ylabel="Diameter of intense \n chromatin (m)")
axes[0].set(xlabel=None)

pairs = [(labels[0], labels[1])]

annotator = Annotator(ax=axes[0], pairs=pairs, data=df, x="Condition", y=feature, order=labels)
annotator.configure(test="Mann-Whitney", text_format="star", loc="outside", text_offset=-2)
annotator.apply_and_annotate()

feature = "Number of total intense chromatin"
sns.boxplot(
    data=df_total,
    x="Condition",
    y=feature,
    order=labels,
    # palette=my_pal,
    color="white",
    width=0.5,
    boxprops=dict(edgecolor="black"),
    capprops=dict(color="black"),
    whiskerprops=dict(color="black"),
    medianprops=dict(color="black"),
    # flierprops=dict(marker='o', markersize=4, markerfacecolor='white', markeredgecolor='white'), # outlier properties
    ax=axes[1],
)

# Add swarmplot for dots in the boxplot
sns.swarmplot(data=df_total, x="Condition", y=feature, order=labels, palette=my_pal, size=5, ax=axes[1])

axes[1].set(ylabel="Number of intense \n chromatin regions")
axes[1].set(xlabel=None)

# Statistics
pairs = [(labels[0], labels[1])]  # Compare non-confined and confined

annotator = Annotator(data=df_total, x="Condition", y=feature, pairs=pairs, order=labels, ax=axes[1])
annotator.configure(test="Mann-Whitney", text_format="star", loc="outside", text_offset=-2)
annotator.apply_and_annotate()

lw = 1.5

axes[0].spines["top"].set_visible(False)
axes[0].spines["right"].set_visible(False)
axes[0].spines["left"].set_linewidth(lw)
axes[0].spines["bottom"].set_linewidth(lw)
axes[0].tick_params(axis="both", which="major", width=lw)

axes[1].spines["top"].set_visible(False)
axes[1].spines["right"].set_visible(False)
axes[1].spines["left"].set_linewidth(lw)
axes[1].spines["bottom"].set_linewidth(lw)
axes[1].tick_params(axis="both", which="major", width=lw)

for p in axes[0].texts:
    p.set_fontsize(fontsize)

for p in axes[1].texts:
    p.set_fontsize(fontsize)


new_confiner_labels_IMR90 = ["Non-\nconfined", "Confined"]

axes[0].set_xticklabels(new_confiner_labels_IMR90)
axes[1].set_xticklabels(new_confiner_labels_IMR90)

plt.savefig("../../Figures/Supplementary/Suppl_Figure5B_C.pdf", bbox_inches="tight", dpi=300)


plt.show()
```
